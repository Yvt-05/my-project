```python
#Inner Merge on common column

import pandas as pd

df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'Name': ['Aman', 'Riya', 'Karan']
})

df2 = pd.DataFrame({
    'id': [2, 3, 4],
    'Marks': [80, 90, 70]
})

result1 = pd.merge(df1, df2, on='id', how='inner')#inner merge on common column
print(result1)
result2 = pd.merge(df1, df2, on='id', how='left')#left merge on common column
print(result2)
```

```python
import pandas as pd

df1 = pd.DataFrame({
    'ID': [1, 2],
    'Name': ['Aman', 'Riya'],
    }, index=[0, 1])

df2 = pd.DataFrame({
    'ID': [2, 3],
    'Marks': [80, 90]
}, index=[0, 1])
result = pd.merge(df1, df2, on='ID', how='right')
print(result)
```

```python
# Concatenate and Merge DataFrames

import pandas as pd

df1 = pd.DataFrame({
    'ID': [1, 2],
    'Name': ['Aman', 'Riya']
})

df2 = pd.DataFrame({
    'ID': [3, 4],
    'Name': ['Karan', 'Simran']
})

df3 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Marks': [80, 90, 85, 95]
})

# Concatenate df1 and df2
new_df = pd.concat([df1, df2])

print("Concatenated DataFrame:")
print(new_df)

# Merge with df3
result = pd.merge(new_df, df3, on='ID')

print("\nMerged DataFrame:")
print(result)

```
