```python
import numpy as np
arr1 = np.array([[3, 4],
                 [5, 6]])
arr2 = np.array([[1, 0],
                 [2, 3]])
# Average of two arrays

avg = (arr1 + arr2) / 2

print("Average of NumPy arrays:\n", avg)


# Mean

mean1 = np.mean(arr1)
mean2 = np.mean(arr2)

print("\nMean of arr1:", mean1)
print("Mean of arr2:", mean2)

# Median

median1 = np.median(arr1)
median2 = np.median(arr2)

print("\nMedian of arr1:", median1)
print("Median of arr2:", median2)

# Mode using NumPy only

# Flatten arrays into 1D
flat1 = arr1.flatten()
flat2 = arr2.flatten()

# Find unique values and counts
values1, counts1 = np.unique(flat1, return_counts=True)
values2, counts2 = np.unique(flat2, return_counts=True)

# Find mode
mode1 = values1[np.argmax(counts1)]
mode2 = values2[np.argmax(counts2)]

print("\nMode of arr1:", mode1)
print("Mode of arr2:", mode2)
```
