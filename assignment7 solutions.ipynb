```python
#Create a Pandas Series from Dictionary
import pandas as pd
data={'a': 20, 'b': 40, 'c': 60}
s=pd.Series(data)
print(s)

#Create a Pandas Series from a list
data1=[1000,2000,3000]
s1=pd.Series(data1)
print(s1)

```

```python
#Access the elements of a Series in Pandas
import pandas as pd
s=pd.Series(['a','b','c','d','e'])
print(s)
print(s[0])
print(s[1])
print(s[2])
print(s[0:2])
```

```python
# Make a Pandas DataFrame with a two-dimensional Python list
import pandas as pd

data = [[1, 'APPLE'], [2, 'BANANA'], [3, 'CHERRY']]

df = pd.DataFrame(data, columns=['quantity', 'Name'])

print(df)
```

```python
# Create DataFrame from Python dict
import pandas as pd

data = {
    'Name': ['Ram', 'Shyam', 'Mohan'],
    'Age': [20, 21, 22]
}

df = pd.DataFrame(data)

print(df)
```

```python
#Different ways to iterate over rows in Pandas DataFrame

import pandas as pd

df = pd.DataFrame({
    'Name': ['Aman', 'Riya', 'Karan'],
    'Marks': [85, 90, 78]
})


for index, row in df.iterrows():
    print("Name:", row['Name'])
    print("Marks:", row['Marks'])
    print()
```

```python
# Selecting rows in pandas DataFrame based on conditions

import pandas as pd


df = pd.DataFrame({
    'City': ['Jaipur', 'Delhi', 'Mumbai', 'Agra'],
    'Temperature': [35, 42, 30, 45]
})

result = df[df['Temperature'] > 40]
print(result)

result1=df[df['City'] == 'Delhi']
print(result1)

result2=df[df['City'].str.contains('a')]
print(result2)

result3=df[df['City'].str.startswith('J')]
print(result3)
#Limited rows selection with given column
print(df.loc[0:1, ['City']])
#Select any row from a DataFrame using iloc[]
print(df.iloc[1])
# Drop rows from dataframe based on condition
df=df[df['Temperature'] <= 40]
print(df)
```

```python
# Insert row at given position in Pandas DataFrame

import pandas as pd

df = pd.DataFrame({
    'City': ['Jaipur', 'Mumbai'],
    'Temperature': [35, 30]
})

df.loc[1] = ['Delhi', 42]
print(df)
```

```python
# Create a list from rows in Pandas dataframe

import pandas as pd

df = pd.DataFrame({
    'City': ['Jaipur', 'Delhi', 'Mumbai', 'Agra'],
    'Temperature': [35, 42, 30, 45]
})


data = df.values.tolist()

print(data)
```
