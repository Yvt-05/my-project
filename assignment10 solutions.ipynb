```python
import numpy as np

arr = np.array([[6, -8, 73, -110], [np.nan, -8, 0, 94]])

arr = np.nan_to_num(arr)

arr = arr[[1, 0], :][:, [1, 0, 2, 3]]
print(arr)
```

```python
import numpy as np

arr = np.random.rand(2, 3, 4)

new_arr = np.moveaxis(arr, 0, 2)
print(new_arr)
```

```python
import numpy as np

arr = np.array([[1, np.nan, 3], [4, 5, np.nan], [np.nan, 7, 9]])
means = np.nanmean(arr, axis=0)
arr[np.isnan(arr)] = np.take(means, np.where(np.isnan(arr))[1])
print(arr)
```

```python
import numpy as np
import pandas as pd

arr = np.array([6, -8, 73, -110, 0, 94])
arr = pd.Series(arr).replace(to_replace=arr[arr < 0], value=0).to_numpy()

print(arr)
```
