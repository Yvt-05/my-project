```python
#question 3
def multiply(l1):
    result = 1
    for x in l1:
        result *= x
    return result

l1=[1, 2, 3, 4, 5]
print(multiply(l1))
```

```python
#question 4
def factorial(n):
    fact = 1

    for i in range(1, n + 1):
        fact = fact * i

    print(fact)

# Input from user
num = int(input("Enter a number: "))

factorial(num)
```

```python
#question 8
def prime(n):
    count = 0
    for i in range(1, n + 1):
        if n % i == 0:
            count += 1
    if count == 2:
        print(n, "is a prime number.")
    else:
        print(n, "is not a prime number.")
num = int(input("Enter a number: "))
prime(num)
   
```

```python
#question o9

s = input("Enter a string: ")

upper = 0
lower = 0

for i in s:
    if i.isupper():
        upper += 1
    elif i.islower():
        lower += 1

print("Upper =", upper)
print("Lower =", lower)
```
