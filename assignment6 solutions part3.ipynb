```python
# https://api.openweathermap.org/data/2.5/weather?q=mumbai&appid=6ffc272d15fc2f5004829b830fb84fce

import requests

def weather_data(url):

    url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid=6ffc272d15fc2f5004829b830fb84fce&units=metric"

    try:
        res = requests.get(url)
        res.raise_for_status()

        data = res.json()

        print("City Name:", data['name'])
        print("Longitude:", data['coord']['lon'])
        print("Latitude:", data['coord']['lat'])
        print("Sea Level:", data['main'].get('sea_level', 'Not Available'))

    except requests.exceptions.HTTPError as e:
        print(e)


city = input("Enter the city name: ")

url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid=6ffc272d15fc2f5004829b830fb84fce&units=metric"

weather_data(url)

# show more details of weather

def weather_details(url):

    try:
        res = requests.get(url)
        res.raise_for_status()

        data = res.json()

        print("Clouds:", data['clouds']['all'])
        print("Sunrise:", data['sys']['sunrise'])
        print("Sunset:", data['sys']['sunset'])
        print("Wind Degree:", data['wind']['deg'])
        print("Timezone:", data['timezone'])

    except requests.exceptions.HTTPError as e:
        print(e)

weather_details(url)
```
