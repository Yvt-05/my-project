```python
nums = (3, 14, 7, 22, 9, 41, 18, 5)
print("sum of all elements:", sum(nums))
print("maximum element:", max(nums))
print("minimum element:", min(nums))
greater_than_10 = [num for num in nums if num > 10]
print("elements greater than 10:", greater_than_10)
```
