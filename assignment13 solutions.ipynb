```python
import numpy as np

arr = np.array([[6, -8, 73, -110],
                [np.nan, -8, 0, 94]])

arr = np.nan_to_num(arr, nan=0)

print("After replacing NaN with 0:")
print(arr)

print("After interchanging rows and columns:")
print(arr.T)
```

```python
import numpy as np

arr = np.array([[1, np.nan, 3],
                [4, 5, np.nan],
                [7, 8, 9]])

col_avg = np.nanmean(arr, axis=0)

inds = np.where(np.isnan(arr))

arr[inds] = np.take(col_avg, inds[1])

print(arr)
```

```python
import numpy as np

arr1 = np.array([[1, 2],
                 [3, 4]])

arr2 = np.array([[2, 3],
                 [4, 5]])

arr = np.concatenate((arr1, arr2))

print("Mean =", np.mean(arr))
print("Median =", np.median(arr))
```

```python
import numpy as np
#x−2y+3z=9
#−x+3y−z=−6
#2x−5y+5z=17

A = np.array([[1,-2,3],
              [-1,3,-1],
              [2,-5,5]])

B = np.array([9,-6,17])

X = np.linalg.inv(A).dot(B)

print("x =", X[0])
print("y =", X[1])
print("z =", X[2])
```
